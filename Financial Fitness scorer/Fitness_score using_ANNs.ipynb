```python
#Importing essential Libraries
import numpy as np
import pandas as pd

#Importing the dataset
train_df = pd.read_csv('Fitness_score_data.csv')

#Dropping unnescessary columns
train_df = train_df.drop(columns=['Customer_ID']) 

#Taking care of missing values for categorical variables by using mode of the values and for numerical values using mean and median of the taking care of missing data.
train_df['Gender'].fillna(train_df['Gender'].mode()[0], inplace=True)
train_df['Marital_status'].fillna(train_df['Marital_status'].mode()[0], inplace=True)
train_df['Number_of_Dependents'].fillna(train_df['Number_of_Dependents'].mode()[0], inplace=True)
train_df['City'].fillna(train_df['City'].mode()[0], inplace=True)
train_df['Type of employment'].fillna(train_df['Type of employment'].mode()[0], inplace=True)
train_df['Credit_score'].fillna(train_df['Credit_score'].mode()[0], inplace=True)
train_df['Property'].fillna(train_df['Property'].mode()[0], inplace=True)
train_df['Education'].fillna(train_df['Education'].mode()[0], inplace=True)
train_df['Loan_term(For the loan user wants)'].fillna(train_df['Loan_term(For the loan user wants)'].mode()[0], inplace=True)
train_df['Loan_amount(That user wants)'].fillna(train_df['Loan_amount(That user wants)'].median(), inplace=True)
train_df['Income'].fillna(train_df['Income'].mean(), inplace=True)
train_df['Meeting_financial_goals'].fillna(train_df['Meeting_financial_goals'].mode()[0], inplace=True)
train_df['Emergency_Funds'].fillna(train_df['Emergency_Funds'].mode()[0], inplace=True)
train_df['Current_Savings'].fillna(train_df['Current_Savings'].mean(), inplace=True)
train_df['Total_loan_amount(At_present)'].fillna(train_df['Total_loan_amount(At_present)'].median(), inplace=True)
train_df['Expenses'].fillna(train_df['Expenses'].mean(), inplace=True)
train_df['Current_number_of_loans'].fillna(train_df['Current_number_of_loans'].median(), inplace=True)
train_df['Number_of_Credit_Cards'].fillna(train_df['Number_of_Credit_Cards'].median(), inplace=True)
train_df['Loan_status(approved or not)'].fillna(train_df['Loan_status(approved or not)'].mode()[0], inplace=True)
train_df['Credit_Card_Approval'].fillna(train_df['Credit_Card_Approval'].mode()[0], inplace=True)
train_df['User_activity'].fillna(train_df['User_activity'].mode()[0], inplace=True)
train_df['Investments(Gold, equity, etc)'].fillna(train_df['Investments(Gold, equity, etc)'].median(), inplace=True)
train_df['EMI'].fillna(train_df['EMI'].mean(), inplace=True)
train_df['Age'].fillna(train_df['Age'].mean(), inplace=True)

#Oversampling the data
X = train_df
from imblearn.over_sampling import RandomOverSampler
rus = RandomOverSampler(random_state=42)
X_train,XE = rus.fit_resample(X, X['Type of employment'])

rus = RandomOverSampler(random_state=42)
X_train,XE = rus.fit_resample(X_train, X_train['Education'])

rus = RandomOverSampler(random_state=42)
X_train,XE = rus.fit_resample(X_train, X_train['Gender'])

rus = RandomOverSampler(random_state=42)
X_train,XE = rus.fit_resample(X_train, X_train['Marital_status'])

rus = RandomOverSampler(random_state=42)
X_train,XE = rus.fit_resample(X_train, X_train['Marital_status'])

rus = RandomOverSampler(random_state=42)
X_train,XE = rus.fit_resample(X_train, X_train['Property'])

rus = RandomOverSampler(random_state=42)
X_train,XE = rus.fit_resample(X_train, X_train['Type of employment'])

rus = RandomOverSampler(random_state=42)
X_train,XE = rus.fit_resample(X_train, X_train['Loan_status(approved or not)'])

rus = RandomOverSampler(random_state=42)
X_train,XE = rus.fit_resample(X_train, X_train['City'])

rus = RandomOverSampler(random_state=42)
X_train,XE = rus.fit_resample(X_train, X_train['Credit_Card_Approval'])

rus = RandomOverSampler(random_state=42)
X_train,XE = rus.fit_resample(X_train, X_train['Emergency_Funds'])

rus = RandomOverSampler(random_state=42)
X_train,XE = rus.fit_resample(X_train, X_train['Meeting_financial_goals'])

#Making Dummy variables
train_df = X_train
train_df = pd.get_dummies(train_df)

#Separating Labels and data for further use
X = train_df.drop(columns = ['Gender_Female', 'Marital_status_Single', 'City_Non-Metro', 'Education_High School', 'Type of employment_Student', 'Property_Semiurban', 'Loan_status(approved or not)_Y', 'Loan_status(approved or not)_N', 'Credit_Card_Approval_Y', 'Credit_Card_Approval_N', 'Meeting_financial_goals_Y', 'Meeting_financial_goals_N', 'Emergency_Funds_Y', 'Emergency_Funds_N'])
yLoan = train_df['Loan_status(approved or not)_Y']
yCred = train_df['Credit_Card_Approval_Y']
yMFG = train_df['Meeting_financial_goals_Y']
yEF = train_df['Emergency_Funds_Y']

#Splitting the data into training and testing sets
from sklearn.model_selection import train_test_split
X1Loan_train, X1Loan_test, y1Loan_train, y1Loan_test = train_test_split(X, yLoan, test_size=0.2, random_state=10)
X1Cred_train, X1Cred_test, y1Cred_train, y1Cred_test = train_test_split(X, yCred, test_size=0.2, random_state=10)
X1MFG_train, X1MFG_test, y1MFG_train, y1MFG_test = train_test_split(X, yMFG, test_size=0.2, random_state=10)
X1EF_train, X1EF_test, y1EF_train, y1EF_test = train_test_split(X, yEF, test_size=0.2, random_state=10)
```

```python
#Implementing Artificial Neural Networks
import keras
from keras.models import Sequential
from keras.layers import Dense, Dropout
from sklearn.model_selection import cross_val_score,GridSearchCV
from keras.wrappers.scikit_learn import KerasClassifier

classifier_ANN_Loan= Sequential()
classifier_ANN_Loan.add(Dense(output_dim=20, init ='uniform', activation='relu', input_dim=25))
classifier_ANN_Loan.add(Dropout(p=0.1))
classifier_ANN_Loan.add(Dense(output_dim=10, init ='uniform', activation='relu'))
classifier_ANN_Loan.add(Dropout(p=0.1))
classifier_ANN_Loan.add(Dense(output_dim=1, init ='uniform', activation='sigmoid'))
classifier_ANN_Loan.compile(optimizer='Adam', loss='binary_crossentropy', metrics=['accuracy'])
classifier_ANN_Loan.fit(X1Loan_train,y1Loan_train, batch_size=1, nb_epoch=25)

classifier_ANN_Cred= Sequential()
classifier_ANN_Cred.add(Dense(output_dim=20, init ='uniform', activation='relu', input_dim=25))
classifier_ANN_Cred.add(Dropout(p=0.1))
classifier_ANN_Cred.add(Dense(output_dim=10, init ='uniform', activation='relu'))
classifier_ANN_Cred.add(Dropout(p=0.1))
classifier_ANN_Cred.add(Dense(output_dim=1, init ='uniform', activation='sigmoid'))
classifier_ANN_Cred.compile(optimizer='Adam', loss='binary_crossentropy', metrics=['accuracy'])
classifier_ANN_Cred.fit(X1Cred_train,y1Cred_train, batch_size=1, nb_epoch=20)

classifier_ANN_MFG= Sequential()
classifier_ANN_MFG.add(Dense(output_dim=20, init ='uniform', activation='relu', input_dim=25))
classifier_ANN_MFG.add(Dropout(p=0.1))
classifier_ANN_MFG.add(Dense(output_dim=10, init ='uniform', activation='relu'))
classifier_ANN_MFG.add(Dropout(p=0.1))
classifier_ANN_MFG.add(Dense(output_dim=1, init ='uniform', activation='sigmoid'))
classifier_ANN_MFG.compile(optimizer='Adam', loss='binary_crossentropy', metrics=['accuracy'])
classifier_ANN_MFG.fit(X1MFG_train,y1MFG_train, batch_size=1, nb_epoch=25)

classifier_ANN_EF= Sequential()
classifier_ANN_EF.add(Dense(output_dim=20, init ='uniform', activation='relu', input_dim=25))
classifier_ANN_EF.add(Dropout(p=0.1))
classifier_ANN_EF.add(Dense(output_dim=10, init ='uniform', activation='relu'))
classifier_ANN_EF.add(Dropout(p=0.1))
classifier_ANN_EF.add(Dense(output_dim=1, init ='uniform', activation='sigmoid'))
classifier_ANN_EF.compile(optimizer='Adam', loss='binary_crossentropy', metrics=['accuracy'])
classifier_ANN_EF.fit(X1EF_train,y1EF_train, batch_size=1, nb_epoch=20)
```
